```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Replace 'data.csv' with the path to your CSV file
df_current = pd.read_csv('current/log.csv')
# Another example with customizations
plt.figure(figsize=(10, 6))
plt.plot(df_current['time'], df_current['XgA'], linestyle='-', color='red', label='XgA', linewidth=1)
plt.plot(df_current['time'], df_current['XgB'], linestyle='-', color='blue', label='XgB', linewidth=1)
plt.title('Current')
plt.xlabel('time')
plt.ylabel('mass fractions')
plt.legend()
plt.grid(True)
plt.show()
```
